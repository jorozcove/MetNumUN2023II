```python
import numpy as np
import matplotlib.pyplot as plt

def cubic(b,c,d,x):
    """Calulate the value of the  qubic equation, v = x**3 + bx**2 + c*x +d

    Parameters
    ----------
    b, c,d  : float
       Coefficients

    Returns
    -------
    v : float """

    # ... enter your code here ...
    v = x**3 + b*x**2 + c*x +d
    return v



x=np.linspace(-10,10,100)
y = cubic(2,3,-2,x)
y2 = cubic(5,3,9,x)
y3 = cubic(1,1,1,x)
plt.plot(x, y, color="green")
plt.plot(x, y2, color="red")
plt.plot(x, y3, color="blue")
plt.show()
```

```python
import math
import numpy as np

def solve_cubic(b, c, d):

    """Solve a cubic equation, x**3 + bx**2 + c*x +d = 0.

    Parameters
    ----------
    b, c,d : float
       Coefficients

    Returns
    -------
    x1, x2,x3 : float or complex
       Roots.  """

    # ... enter your code here ...

    f = discriminantCase1(b, c)
    g = discriminantCase2(b, c, d)
    h = discriminantCase3(g, f)

    if f == 0 and g == 0 and h == 0:
        if d >= 0:
            x = (d) ** (1 / 3.0) * -1
        else:
            x = (-d) ** (1 / 3.0)
        return np.array([x, x, x])

    elif h <= 0:

        i = math.sqrt(((g ** 2.0) / 4.0) - h)
        j = i ** (1 / 3.0)
        k = math.acos(-(g / (2 * i)))
        L = j * -1
        M = math.cos(k / 3.0)
        N = math.sqrt(3) * math.sin(k / 3.0)
        P = (b / (3.0)) * -1

        x1 = 2 * j * math.cos(k / 3.0) - (b / (3.0))
        x2 = L * (M + N) + P
        x3 = L * (M - N) + P

        return x1,x2,x3

    elif h > 0:
        R = -(g / 2.0) + math.sqrt(h)
        if R >= 0:
            S = R ** (1 / 3.0)
        else:
            S = (-R) ** (1 / 3.0) * -1
        T = -(g / 2.0) - math.sqrt(h)
        if T >= 0:
            U = (T ** (1 / 3.0))
        else:
            U = ((-T) ** (1 / 3.0)) * -1

        x1 = (S + U) - (b / (3.0))
        x2 = -(S + U) / 2 - (b / (3.0)) + (S - U) * math.sqrt(3) * 0.5j
        x3 = -(S + U) / 2 - (b / (3.0)) - (S - U) * math.sqrt(3) * 0.5j

        return x1,x2,x3



def discriminantCase1(b, c):
    return ((3.0 * c) - ((b ** 2.0))) / 3.0


def discriminantCase2(b, c, d):
    return (((2.0 * (b ** 3.0))) - ((9.0 * b * c)) + (27.0 * d)) /27.0


def discriminantCase3(g, f):
    return ((g ** 2.0) / 4.0 + (f ** 3.0) / 27.0)


b = 1.0
c = -6.0
d = 11.0
roots = solve_cubic(b, c, d)
print("Roots:", roots)
```

Method used: Cardano

El método de Cardano es un método algebraico destinado para resolver analíticamente cualquier ecuación cúbica y que apareció por primera vez en el libro Ars Magna en 1545 publicado por el matemático italiano Gerolamo Cardano (1501-1576).<br>
Cardano resuelve ecuaciones cúbicas $ax^3 + bx^2 + cx + d = 0$ mediante la sustitución $x = u - \frac{b}{3a}$ para simplificarla. Luego introduce $v$, resuelve para $u$ y $v$, y expresa las raíces $x$ en términos de $u$ y $v$. Se abordan casos de discriminante para raíces reales o complejas. El método involucra pasos algebraicos complejos. Finalmente, se sustituyen $u$ y $v$ para obtener las raíces de la ecuación original. Esta técnica histórica combina sustitución y manipulación algebraica para ofrecer una solución completa a ecuaciones cúbicas.
